Пропуски в столбцах `eu_sales` и `jp_sales` появились в результате преобразования типа данных в этих столбцах. Заменим пропуски на среднее значение в зависимости от названия платформы и года выхода игры.

- Обработайте пропущенные значения. Для каждого случая вы можете выбрать оптимальный, на ваш взгляд, вариант: заменить на определённое значение, оставить как есть или удалить.
- Если вы решите заменить пропуски на значение-индикатор, то убедитесь, что предложенное значение не может быть использовано в данных.
- Если вы нашли пропуски в данных с количеством проданных копий игры в том или ином регионе, их можно заменить на среднее значение в зависимости от названия платформы и года выхода игры.

```python
# Перед обработкой пропусков сохраним исходное количество строк
first_row_count = df.shape[0]
```

Заменим пропуски в данных с количеством проданных копий игры на среднее значение в зависимости от названия платформы и года выхода игры:

```python
def sales_count(row):
    if pd.isna(row['eu_sales']):
        group = df[(df['platform'] == row['platform']) &
                   (df['year_of_release'] == row['year_of_release'])]
        return group['eu_sales'].mean()
    else:
        return row['eu_sales']

df['eu_sales'] = df.apply(sales_count, axis=1)
```

```python
def sales_count(row):
    if pd.isna(row['jp_sales']):
        group = df[(df['platform'] == row['platform']) &
                   (df['year_of_release'] == row['year_of_release'])]
        return group['jp_sales'].mean()
    else:
        return row['jp_sales']

df['jp_sales'] = df.apply(sales_count, axis=1)
```

Удалим из датафрейма строки с пропусками в столбцах `name`, `genre` и `year_of_release`:

```python
df = df.dropna(subset = ['name', 'genre', 'year_of_release'])
```

## 4. Категоризация данных
    
Проведите категоризацию данных:
- Разделите все игры по оценкам пользователей и выделите такие категории: высокая оценка (от 8 до 10 включительно), средняя оценка (от 3 до 8, не включая правую границу интервала) и низкая оценка (от 0 до 3, не включая правую границу интервала).

Столбцы `user_score` и `critic_score` содержат большое количество пропусков (53,8 и 50,7 % соответственно), поскольку удаление данных приведет к значительной потере данных, а заполнение данных средним значением может значительно исказить картину (оценка игр одинаковых жанров и выпущенных на одинаковых платформах не всегда являются одинаковыми. Даже разные части одной и той же серии игр могут иметь сильно разные оценки) для категоризации данных выделим четвертую категорию `нет оценки`, а результаты будем оценивать среди игр, у которых имеются оценки критиков и игроков.

```python
df_actual = df_actual.copy()
```

```python
def categorize_user_score(row):
    if pd.isna(row['user_score']):
        return 'нет оценки'
    elif row['user_score'] >= 8:
        return 'высокая оценка'
    elif row['user_score'] >= 3:
        return 'средняя оценка'
    else:
        return 'низкая оценка'

df_actual['user_category'] = df_actual.apply(categorize_user_score, axis=1)
```

- Разделите все игры по оценкам критиков и выделите такие категории: высокая оценка (от 80 до 100 включительно), средняя оценка (от 30 до 80, не включая правую границу интервала) и низкая оценка (от 0 до 30, не включая правую границу интервала).

```python
def categorize_critic_score(row):
    if pd.isna(row['critic_score']):
        return 'нет оценки'
    elif row['critic_score'] >= 80:
        return 'высокая оценка'
    elif row['critic_score'] >= 30:
        return 'средняя оценка'
    else:
        return 'низкая оценка'

df_actual['critic_category'] = df_actual.apply(categorize_critic_score, axis=1)
```

```python
user_category_count = df_actual['user_category'].value_counts()
display(user_category_count)
```

- `user_category` - категория игры на основе оценок пользователей;  
- `critic_category` - категория игры на основе оценок критиков;  

Анализируя результаты проведенного ислледования можно выделить следующие основные моменты: 
- За период с 2000 по 2013 год было выпущено 12781 игра различных жанров и для разных платформ;
- Наиболее популярными плаформами являются: PS2, DS, Wii, PSP И X360;
- Согласно проведенной категоризации большая часть игр (без учета игр у которых отсутствуют данные об оценках пользователей и критиков) оценивалась по среднему уровню и критиками и пользователями
